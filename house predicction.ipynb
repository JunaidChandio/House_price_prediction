```python
num_values=df.isnull().sum()/df.shape[0]*100
```

```python
num_values
```

```python
num_values_per_g_0=num_values[num_values>0].keys()
```

```python
num_values_per_g_0
```

```python
df2=df[num_values_per_g_0].fillna(df[num_values_per_g_0].mean())
```

```python
df2.isnull()
```

```python
df2.isnull().sum()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df2.isnull())
plt.title("this heatmap will show null values in dataframe if any ")
```

```python
# split dataframe

X=df2.iloc[:,0:-1]
y=df2.iloc[:,-1]
```

```python
X.isnull().sum()
```

```python
y.isnull().sum()
```

```python
# train dataset

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(X_train)
```

```python
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train_sc,y_train)
```

```python
regressor.score(X_test_sc,y_test)
```

```python
pred=regressor.predict(X_test_sc)
pred
```

```python
from sklearn.linear_model import Ridge
```

```python
lr=Ridge()
```

```python
lr.fit(X_train_sc,y_train)
```

```python
lr.score(X_test_sc,y_test)
```

```python
from sklearn.linear_model import Lasso
```

```python
ls=Lasso()
```

```python
ls.fit(X_train_sc,y_train)
```

```python
ls.score(X_test_sc,y_test)
```

```python
ls_pred=ls.predict(X_test_sc)
ls_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,ls_pred)
```

```python
rmse=np.sqrt(mse)
```

```python
print("mse",mse)
print("rmse",rmse)
```
